__Importar librerías__

```python
import pandas as pd
import matplotlib.pyplot as plt
```

__Importar el dataframe__

```python
archivo_excel = "Datos Desafio 3.xlsx"
data = pd.read_excel(archivo_excel)
```

__Visualizar el dataframe__

```python
print(data.head())
```

__Eliminar filas/columnas con valores faltantes__

```python
datos_limpios = data.dropna()
```

__Visulizar los datos limpios__

```python
print(datos_limpios.head())
```

__Se genera un dataframe__

```python
df = pd.DataFrame(datos_limpios)
```

__Se calcula la media del espesor del lomo, la desviación estándar del porcentaje de magro y el coeficiente de correlación entre el peso de la canal caliente y el peso de la canal refrigerada__

```python
media_lomo = df['Espesor Lomo\n  (m.m)'].mean()
desviacion_estandar_cm = df['Porcentaje Magro\n CM (%)'].std()
correlacion = df['Peso Canal Caliente \n (PCC- Kg)'].corr(df['Peso Canal Refrigerada (PCR - Kg)'])
```

__Se genera un histograma del espesor del lomo__

```python
df['Espesor Lomo\n  (m.m)'].plot.hist(bins=5)
plt.xlabel('Espesor Lomo\n  (m.m)')
plt.ylabel('Frecuencia')
plt.title('Distribución del Espesor del Lomo')
plt.show()
```

__Se genera un diagrama de dispersión para mostrar la relación entre el peso de la canal caliente y el peso de la canal refrigerada__

```python
df.plot.scatter(x='Peso Canal Caliente \n (PCC- Kg)', y='Peso Canal Refrigerada (PCR - Kg)')
plt.xlabel('Peso Canal Caliente (PCC- Kg)')
plt.ylabel('Peso Canal Refrigerada (PCR - Kg)')
plt.title('Relación entre Peso Canal Caliente y Peso Canal Refrigerada')
plt.show()
```
